# Selecting with iloc and loc

_iloc :-_ This functions always accept integral parameters for functioning

_loc :-_ This function accepts both numerical and textual parameters for functioning

```python
data.iloc[1] # fetching the 1st index or the 2nd row

# You can see here that the columns are visualized as row in the output
```

```python
data.iloc[1,[1,2,3]] # fetching the 2nd row and 2nd , 3rd, 4th column
```

```python
data.loc["Paris",["one","two"]] # access the paris index and 'one' and 'two' columns
```

# Reindexing and Dropping

```python
df=pd.DataFrame(np.arange(9).reshape(3,3),
                index=["a","c","d"],
                columns=["Tim","Tom","Kate"])
df
```

```python
df2=df.reindex(["d","c","b","a"]) # indices will be renamed with the provided index.
'''
If there exists an index in the reindexing which was already present in the original data , reindexing will preserve that
row with the same index. If new indices are more than the availability of columns in the original data , then rest indices
will NaN type.
'''
df2
```

```python
names=["Kate","Tim","Tom"]
'''
You can even change or arange the columns using the indexing method. If you use the same name for idexing again , it will
change the order of the new display as per the order you entered and preserve the data in the same column name. Again you
must remember that this display will be valid for this cell only.
'''
df.reindex(columns=names)
```

```python
s=pd.Series(np.arange(5.),
            index=["a","b","c","d","e"])
s
```

```python
new_s=s.drop("b") # we are dropping a row here. There is a default parameter hidden , always set as default to rows called
# axis . This determines that wether you are operating on row or column and 0 always means row.
new_s
```

# Working with the missing data

```python
s.isnull() # checks if any column have any null value or not
```

```python
s.notnull() # vice versa of is null
```

```python
s[3]=None # you can assign any value as null in this way as well.
s.isnull()
```

```python
s.dropna() # drops the rows with atleast one none or NaN values in them
```

```python
from numpy import nan as NA # you can even give alias to nan
```

```python
df=pd.DataFrame([[1,2,3],[4,NA,5],
                 [NA,NA,NA]])
df
```

```python
df.dropna()
```

```python
df.dropna(how="all") # you can set conditions that when to drop a column . Like in this case it drops the row when all the
# elements in that row are nan , then it will drop that row

# by default how is set to 'any'
```

```python
df[1]=NA
df
```

```python
df.dropna(axis=1,how="all") # changing the orientation to column operation
```

```python
df # you can see that the elements in the df are still intact , so the operations done previously were stored in a new location
```

```python
df.fillna(0) # This will fill all the nans with 0
```

```python
df=pd.DataFrame([[1,2,3],[4,NA,5],
                 [NA,NA,NA]])
df
```

```python
df.fillna(method="ffill") # this will all the nans with values in the previous cell of the nan in the same column
```

```python
df.fillna(method="ffill",limit=1) #this will limit to fill nan only at one place in each column.
```

```python
df.fillna(df.mean()) # this will fill the nans with mean of that column . The means are claculated excluding the nan values.
```

# What is Groupby in Pandas?

```python
df=pd.DataFrame(
    {"key1":list("aabbab"),
     "key2":["one","two","three"]*2,
     "data1":np.random.randn(6),
     "data2":np.random.randn(6)})
df
```

```python
group=df["data1"].groupby(df["key1"]) #data1 column is now stored in another variable and grouped by Key1
print(group)
```

```python
group.mean()
```

```python
ave=df["data1"].groupby([df["key1"],
                         df["key2"]]).mean() # this will help to accumulate mean of the data1 grouped by key1 and key2
ave
```
