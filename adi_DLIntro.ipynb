## Basic terminologies in deep learning

- **Neural Network**: A computational model inspired by the structure and function of the human brain, consisting of interconnected nodes (neurons) organized in layers.

- **Artificial Neural Network (ANN)**: A computational model composed of layers of interconnected artificial neurons, used to learn complex patterns in data.

- **Layer**: A collection of neurons in a neural network that process input data. Layers are typically organized into input, hidden, and output layers.

- **Input Layer**: The first layer of a neural network, responsible for receiving input data.

- **Hidden Layer**: Intermediate layers in a neural network between the input and output layers, where computations are performed.

- **Output Layer**: The final layer of a neural network, responsible for producing output predictions.

- **Activation Function**: A mathematical function applied to the output of each neuron in a neural network, introducing nonlinearity and enabling the network to learn complex patterns.

- **Backpropagation**: A training algorithm used to update the weights of a neural network by propagating the error backward from the output layer to the input layer.

- **Gradient Descent**: An optimization algorithm used to minimize the error (loss) of a neural network by iteratively adjusting the weights based on the gradient of the loss function with respect to the weights.

- **Loss Function**: A function that measures the difference between the predicted output of a neural network and the true output, used to guide the training process.

- **Batch**: A subset of training examples used in each iteration of the training process.

- **Epoch**: One complete pass through the entire training dataset during the training process.

- **Learning Rate**: A hyperparameter that controls the size of the step taken during gradient descent, influencing the speed and stability of the training process.

- **Overfitting**: A phenomenon where a model learns to perform well on the training data but fails to generalize to unseen data, due to capturing noise or irrelevant patterns in the training data.

- **Underfitting**: A phenomenon where a model is too simple to capture the underlying structure of the data, resulting in poor performance on both the training and test data.

These are some of the basic terminologies commonly used in deep learning. Familiarity with these terms is essential for understanding and working with deep learning models effectively.

## Neurons

**1. Neurons in Artificial Neural Networks (ANNs):**

In ANNs, a neuron is a basic computational unit inspired by the biological neurons in the human brain. Each neuron receives input signals, processes them, and produces an output signal. Here's how a neuron typically works mathematically:

**Mathematical Aspects:**

- **Input Signals**: Neurons in ANNs receive input signals from other neurons or directly from the input data. These input signals are represented as a vector $ \mathbf{x} $ where each element corresponds to a feature or input value.

- **Weights**: Each input signal is associated with a weight $ w_i $, which determines the importance of that input signal. The weights are represented as a vector $ \mathbf{w} $.

- **Bias**: Additionally, a neuron typically includes a bias term $ b $, which allows the neuron to shift its activation function horizontally.

- **Activation Function**: The weighted sum of the input signals along with the bias is passed through an activation function $ f $ to produce the output of the neuron. The activation function introduces non-linearity into the model, allowing it to learn complex patterns. Common activation functions include the sigmoid function, hyperbolic tangent (tanh) function, ReLU (Rectified Linear Unit) function, and softmax function.

- **Output**: The output $ y $ of the neuron is calculated as follows:
  
  $ y = f(\sum_{i=1}^{n} w_i x_i + b) $

**Practical Aspects:**

- **Training**: During the training process, the weights and biases of the neurons are adjusted iteratively using optimization algorithms such as gradient descent, aiming to minimize the difference between the predicted outputs and the actual outputs (target values) for a given set of input data.

- **Backpropagation**: The adjustment of weights and biases is typically done through backpropagation, where the error signal is propagated backward through the network, and the gradients of the error with respect to the weights and biases are computed. These gradients are then used to update the parameters in the direction that minimizes the error.

- **Hyperparameters**: Neurons have associated hyperparameters such as the type of activation function used, the initial values of weights and biases, learning rate, etc., which need to be carefully chosen to ensure optimal performance of the network.

- **Model Complexity**: The number of neurons in a network and the architecture (e.g., number of layers, connectivity between neurons) determine the model's complexity and capacity to learn. Balancing model complexity is crucial to avoid overfitting or underfitting.

- **Practical Activation Functions**: While there are several activation functions, ReLU is commonly used in hidden layers due to its simplicity and effectiveness in training deep networks. Sigmoid and softmax functions are often used in output layers for binary and multi-class classification tasks, respectively.



## Activation Functions

Activation functions are mathematical operations applied to the output of each neuron in a neural network. They introduce non-linearity to the network, allowing it to model complex relationships in data. They help in determining the output of a neural network, as well as its ability to learn and approximate any function.

Some common activation functions include:

1. **Sigmoid**: Outputs values between 0 and 1, often used in binary classification tasks.

2. **ReLU (Rectified Linear Unit)**: Outputs the input directly if it is positive, otherwise outputs zero. It's widely used due to its simplicity and effectiveness.

3. **Tanh (Hyperbolic Tangent)**: Similar to the sigmoid function but outputs values between -1 and 1. It's often used in hidden layers of neural networks.

4. **Softmax**: Used in the output layer for multi-class classification tasks, it converts raw scores into probabilities.


## The Sigmoid function

The sigmoid function, also known as the logistic function, is a widely used activation function in neural networks. It squashes the input values between 0 and 1, which makes it particularly useful in binary classification tasks where the output needs to represent probabilities.

Mathematically, the sigmoid function is defined as:

$ \sigma(z) = \frac{1}{1 + e^{-z}} $

Where:
- $ z $ is the input to the function.

The sigmoid function has the following properties:
1. **Range**: The output of the sigmoid function always lies between 0 and 1.
2. **Smoothness**: The sigmoid function is smooth and continuously differentiable, making it suitable for gradient-based optimization algorithms like backpropagation.
3. **Monotonicity**: The sigmoid function is monotonically increasing, which means as the input increases, the output also increases, but it saturates as the input becomes very large or very small.

## Limitation in Vanishing gradient

When training a neural network using techniques like backpropagation and gradient descent, the network learns by adjusting the weights based on the gradients of the loss function with respect to those weights. However, in deep networks with many layers, gradients tend to get smaller and smaller as they propagate backward through the network during the backpropagation process. This happens due to the chain rule of calculus, which involves multiplying many gradients together. As a result, the gradients can vanish, becoming very close to zero.

1. **Saturation**: The sigmoid function saturates when its input is very large or very small. This means that for extreme input values (either very positive or very negative), the slope of the sigmoid function becomes almost flat, approaching zero. As a result, the gradients of the sigmoid function become very small in these regions.

2. **Derivative of the Sigmoid function**: The derivative of the sigmoid function, which is used in backpropagation to compute gradients, has a maximum value of 0.25. This means that the magnitude of the gradients can decrease rapidly with each layer during backpropagation, especially in deep networks where there are many layers. As gradients are multiplied together during backpropagation, the vanishing gradient problem can be exacerbated, leading to slow learning or even stagnation in the training process.

3. **Limited Range**: The sigmoid function maps its input to the range [0, 1]. As a result, when using the sigmoid activation function in deep networks, the output of each layer tends to be confined to a narrow range, making it challenging for subsequent layers to learn meaningful representations of the data. This limitation can hinder the overall learning process, especially in deeper networks.

Due to these reasons, the use of sigmoid activation functions in deep neural networks has decreased in favor of other activation functions like ReLU (Rectified Linear Unit) and its variants, which do not suffer from the vanishing gradient problem to the same extent


```python
import numpy as np

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Example usage
z = np.random.randint(100)
print("Sigmoid of", z, ":", sigmoid(z))
```

```python
import matplotlib.pyplot as plt

# Generate input values
x = np.linspace(-10, 10, 100)

# Compute sigmoid values for the input range
sigmoid_values = sigmoid(x)

# Plot sigmoid function
plt.figure(figsize=(8, 6))
plt.plot(x, sigmoid_values, label='Sigmoid Function')
plt.title('Sigmoid Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.legend()
plt.show()

# The resulting plot illustrates the sigmoid function's characteristic S-shaped curve, where the output
# values gradually increase from 0 to 1 as the input values vary. Additionally, it shows how the sigmoid
# function maps any real-valued input to a bounded output between 0 and 1, making it suitable for
# binary classification tasks.
```

## The Tanh function

The hyperbolic tangent (tanh) function is another popular activation function used in neural networks. Similar to the sigmoid function, it squashes the input values to a range between -1 and 1, but it is symmetric around the origin, which allows it to model both positive and negative input values.

Mathematically, the tanh function is defined as:

$ \text{tanh}(z) = \frac{e^z - e^{-z}}{e^z + e^{-z}} $

Where:
- $ z $ is the input to the function.

Now, let's derive the tanh function:

1. **Definition**:
$ \text{tanh}(z) = \frac{e^z - e^{-z}}{e^z + e^{-z}} $

2. **Expanding**:
$ \text{tanh}(z) = \frac{(e^z + e^{-z}) - 2e^{-z}}{e^z + e^{-z}}$

3. **Simplifying**:
$ \text{tanh}(z) = 1 - \frac{2e^{-z}}{e^z + e^{-z}} $

4. **Using the identity** $ \frac{e^{-z}}{e^z + e^{-z}} = \frac{1}{e^{2z} + 1} $:
$ \text{tanh}(z) = 1 - \frac{2}{e^{2z} + 1}$

Now, let's discuss some properties of the tanh function:

1. **Range**: The output of the tanh function is between -1 and 1, which makes it zero-centered and allows it to model both positive and negative input values.

2. **Smoothness**: Like the sigmoid function, the tanh function is smooth and continuously differentiable, making it suitable for gradient-based optimization algorithms.

3. **Monotonicity**: The tanh function is monotonic, meaning that it preserves the order of the input values.

##Limitations

1. **Vanishing Gradient**: Like the sigmoid function, the tanh function also suffers from the vanishing gradient problem. As the input values become very large or very small, the gradients of the tanh function approach zero, which can hinder the training process, especially in deep networks.

2. **Saturated Regions**: The tanh function saturates for large positive and negative inputs, leading to very small gradients in these regions. This can slow down the learning process, particularly during the initial stages of training when the weights are randomly initialized.

3. **Zero-Centered**: While being zero-centered can be advantageous for optimization, it also means that the outputs of the tanh function are centered around zero, which might not always be desirable depending on the nature of the data.

Overall, while the tanh function has some desirable properties such as zero-centeredness and a bounded output range, its limitations in terms of the vanishing gradient problem and saturated regions have led to the widespread adoption of activation functions like ReLU (Rectified Linear Unit) and its variants in modern neural networks.

```python
def tanh(z):

    return (np.exp(z) - np.exp(-z)) / (np.exp(z) + np.exp(-z))

# Example usage
z = np.random.randint(100)
print("Tanh of", z, ":", tanh(z))
```

```python
# Generate input values
x = np.linspace(-10, 10, 100)

# Compute tanh values for the input range
tanh_values = tanh(x)

# Plot tanh function
plt.figure(figsize=(8, 6))
plt.plot(x, tanh_values, label='Tanh Function')
plt.title('Tanh Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.legend()
plt.show()
```

## The ReLU function

The Rectified Linear Unit (ReLU) activation function is one of the most widely used activation functions in modern neural networks. It introduces non-linearity by outputting the input directly if it is positive, and zero otherwise. The ReLU function is defined mathematically as:

$ \text{ReLU}(z) = \max(0, z) $

Where:
- $ z $ is the input to the function.

Now, let's discuss some properties of the ReLU function:

1. **Non-Linearity**: Although simple in nature, the ReLU function introduces non-linearity to the network, allowing it to learn complex relationships in the data.

2. **Sparsity**: ReLU neurons can be sparsely activated, meaning that only a subset of neurons is activated for any given input. This can help with computational efficiency and regularization.

3. **Efficiency**: The ReLU function is computationally efficient to compute and differentiable almost everywhere, which facilitates faster training compared to functions like the sigmoid or tanh.

Now, let's derive the ReLU function:

1. **Definition**:
$ \text{ReLU}(z) = \max(0, z) $


## Limitations

1. **Dying ReLU Problem**: ReLU neurons can sometimes become "dead" during training, where they output zero for all inputs. This happens when the input to the ReLU function is negative, causing the neuron to remain inactive and not update its weights during training. This issue can lead to a significant portion of the network being inactive and affects the model's capacity to learn.

2. **Unbounded Output**: The ReLU function does not have an upper bound on its output, which can lead to exploding gradients during training, especially in deep networks. This issue can destabilize the training process and make it difficult to optimize the network.

3. **Not Zero-Centered**: Unlike activation functions like tanh and sigmoid, the ReLU function is not zero-centered. This can lead to undesirable behaviors, such as gradients flowing only in one direction during backpropagation, which can slow down the training process.

Despite these limitations, the ReLU function remains one of the most popular choices for activation functions in deep neural networks due to its simplicity, efficiency, and effectiveness in many practical applications. Several variants of ReLU, such as Leaky ReLU, Parametric ReLU (PReLU), and Exponential Linear Unit (ELU), have been proposed to address some of its limitations while retaining its advantages.

```python
def relu(z):

    return np.maximum(0, z)

# Example usage
z = np.random.randint(100)
print("ReLU of", z, ":", relu(z))
```

```python
# Generate input values
x = np.linspace(-10, 10, 100)

# Compute ReLU values for the input range
relu_values = relu(x)

# Plot ReLU function
plt.figure(figsize=(8, 6))
plt.plot(x, relu_values, label='ReLU Function')
plt.title('ReLU Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.legend()
plt.show()
```

## The softmax function

The softmax function is commonly used as the activation function in the output layer of neural networks, particularly in multi-class classification tasks. It takes as input a vector of real numbers and normalizes it into a probability distribution consisting of probabilities proportional to the exponentials of the input values. This ensures that the output values sum up to 1, making them interpretable as probabilities.

Mathematically, the softmax function is defined as:

$ \text{softmax}(\mathbf{z})_i = \frac{e^{z_i}}{\sum_{j=1}^{K} e^{z_j}} $

Where:
- $ \mathbf{z} $ is the input vector of real numbers.
- $ z_i $ is the $ i $-th element of the input vector.
- $ K $ is the number of classes.

Now, let's discuss some properties of the softmax function:

1. **Output as Probability Distribution**: The softmax function outputs a probability distribution over the $ K $ classes, with each element representing the probability of the corresponding class.

2. **Normalization**: The softmax function normalizes the input values such that the sum of the output probabilities equals 1, ensuring that they can be interpreted as probabilities.

3. **Exponentiation**: The softmax function exponentiates the input values, which amplifies the differences between the input values, making the probabilities more distinguishable.

Now, let's derive the softmax function:

1. **Definition**:
$ \text{softmax}(\mathbf{z})_i = \frac{e^{z_i}}{\sum_{j=1}^{K} e^{z_j}} $

2. **Normalization**:
$ \sum_{i=1}^{K} \text{softmax}(\mathbf{z})_i = \sum_{i=1}^{K} \frac{e^{z_i}}{\sum_{j=1}^{K} e^{z_j}} = 1 $



Now, let's discuss some limitations of the softmax function:

1. **Numerical Stability**: When the input values to the softmax function are large, the exponentiation operation can lead to numerical overflow issues. Similarly, when the input values are very small, exponentiating them can lead to numerical underflow issues. These issues can affect the numerical stability of the computation and the accuracy of the output probabilities.

2. **Sensitivity to Outliers**: The softmax function exponentiates the input values, which amplifies the differences between them. This means that outliers in the input data can disproportionately influence the output probabilities, potentially leading to biased predictions.

3. **Difficulty in Handling Large Number of Classes**: In tasks with a large number of classes, computing the softmax function over all classes can be computationally expensive, as it involves exponentiating and summing over all input values. This can lead to scalability issues, especially in real-time applications or when working with limited computational resources.

Despite these limitations, the softmax function remains a widely used activation function in neural networks, particularly in multi-class classification tasks, due to its interpretability and ability to generate meaningful probability distributions over classes. Various techniques, such as numerical stability improvements and approximate softmax methods, have been proposed to address some of its limitations in practice.

```python
def softmax(z):
    exp_z = np.exp(z)
    return exp_z / np.sum(exp_z)

# Example usage
z = np.array([2.0, 1.0, 0.1])
print("Softmax of", z, ":", softmax(z))
```

```python
x = np.linspace(-10, 10, 100)

# Compute softmax values for the input range
softmax_values = softmax(x)

# Plot softmax function
plt.figure(figsize=(8, 6))
plt.plot(x, softmax_values)
plt.title('Softmax Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.show()
```
