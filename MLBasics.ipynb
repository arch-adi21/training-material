## Introduction to Machine Learning:

### Machine Learning (ML)
is a subset of artificial intelligence that focuses on the development of algorithms enabling computers to learn from data and make intelligent decisions. The primary goal is to allow machines to improve their performance over time without explicit programming.

### Definition and Overview:

Machine Learning involves the use of statistical techniques to enable computers to learn patterns and relationships within data, and then make predictions or decisions based on that learning. It encompasses various algorithms and models that can be classified into different types.

### Historical Context:

The concept of Machine Learning dates back to the 1940s and 1950s, with pioneers like Alan Turing and Arthur Samuel laying the groundwork. However, the real surge in interest and practical applications began in the 21st century, driven by advancements in computational power and the availability of large datasets.

## Types of Machine Learning:

### Supervised Learning:

##### **Definition:**
It involves training a model on a labeled dataset, where the algorithm learns the relationship between input features and corresponding output labels.

Use Cases: Image recognition, speech recognition, spam detection.

Theories: Linear regression, logistic regression, support vector machines.

Mathematical Calculations: Cost functions, gradient descent.

### Unsupervised Learning:

#### **Definition:**

In this type, the model works on unlabeled data, discovering patterns and structures without predefined outputs.

Use Cases: Clustering, dimensionality reduction, anomaly detection.

Theories: K-means clustering, hierarchical clustering, principal component analysis (PCA).

Mathematical Calculations: Euclidean distance, covariance matrices.

### Reinforcement Learning:

#### **Definition:**
Agents learn to make decisions by interacting with an environment. They receive feedback in the form of rewards or penalties.

Use Cases: Game playing (AlphaGo), robotic control, recommendation systems.

Theories: Markov decision processes, Q-learning, policy gradients.

Mathematical Calculations: Bellman equations, value functions.

## **Linear Regression**

**Introduction and Concept:**
Linear Regression is a fundamental statistical method that models the relationship between a dependent variable (target) and one or more independent variables (features) by fitting a linear equation to the observed data. In the context of mAANG interviews, mastering linear regression is crucial, as it forms the basis for more complex machine learning algorithms.

**Mathematical Formulation:**
The basic equation for simple linear regression is:
$$ Y = \beta_0 + \beta_1X + \epsilon  $$
where:
- \( Y \) is the dependent variable,
- \( X \) is the independent variable,
-  $$beta_0$$ is the y-intercept,
- $$beta_1$$ is the slope,
- $$\epsilon$$ represents the error term.

For multiple linear regression with ( n ) independent variables:
$$ Y = \beta_0 + \beta_1X_1 + \beta_2X_2 + ... + \beta_nX_n + \epsilon $$

**Evaluation Metrics:**
1. **Mean Squared Error (MSE):**
   - Definition: MSE measures the average squared difference between the predicted and actual values.
   - Use case: It provides a quantitative measure of how well the model performs.
   - Mathematical Calculation:
     $$MSE = \frac{1}{n} \sum_{i=1}^{n} (Y_i - \hat{Y}_i)^2$$
   - Code Example (Python):
     ```python
     from sklearn.metrics import mean_squared_error
     mse = mean_squared_error(y_true, y_pred)
     ```

2. **R-squared (Coefficient of Determination):**
   - Definition: R-squared represents the proportion of the variance in the dependent variable that is predictable from the independent variables.
   - Use case: Indicates the goodness of fit of the model.

**Practical Examples with Codes:**
Let's consider a real-world scenario where linear regression is applied to predict house prices based on square footage.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1400, 1600, 1700, 1875, 1100, 1550, 2350, 2450, 1425, 1700]).reshape((-1, 1))
y = np.array([245000, 312000, 279000, 308000, 199000, 219000, 405000, 324000, 319000, 255000])

# Model fitting
model = LinearRegression().fit(X, y)

# Prediction
y_pred = model.predict(X)

# Visualization
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='red', linewidth=2)
plt.title('Linear Regression: House Price Prediction')
plt.xlabel('Square Footage')
plt.ylabel('House Price')
plt.show()
```

## Over Fitting of models

Overfitting is a concept in data science, which occurs when a statistical model fits exactly against its training data. When this happens, the algorithm unfortunately cannot perform accurately against unseen data, defeating its purpose. Generalization of a model to new data is ultimately what allows us to use machine learning algorithms every day to make predictions and classify data.

**Solved Question:**
Let's consider a simple Lasso Regression problem:

```python
import numpy as np
from sklearn.linear_model import Lasso

X = np.array([[1, 2], [2, 4], [3, 6], [4, 8]])
```

**Code:**
```python

lasso_model = Lasso(alpha=0.1)

intercept = lasso_model.intercept_

print(f"Coefficients: {coefficients}")
print(f"Intercept: {intercept}")
```

**2. Ridge Regression:**

**Definition:**
Ridge Regression, or L2 regularization, is another linear regression technique that adds a penalty term to the ordinary least squares (OLS) objective function. The penalty is the square of the coefficients' sum multiplied by a regularization parameter $\alpha$.

**Mathematical Formulation:**
The objective function for Ridge Regression is:

**Solved Question:**
Consider the same data as before.

**Code:**
```python
from sklearn.linear_model import Ridge

# Fit Ridge Regression model
ridge_model = Ridge(alpha=0.1)
ridge_model.fit(X, y)

# Coefficients and Intercept
coefficients_ridge = ridge_model.coef_
intercept_ridge = ridge_model.intercept_

print(f"Coefficients: {coefficients_ridge}")
print(f"Intercept: {intercept_ridge}")
```

**Explanation:**
- Both Lasso and Ridge regression are used to prevent overfitting by adding a penalty term to the objective function.
- The regularization parameter $\alpha$ controls the strength of regularization. Higher $\alpha$ values result in stronger regularization.
- Lasso tends to produce sparse models (some coefficients become exactly zero), effectively performing feature selection.
- Ridge, on the other hand, shrinks the coefficients towards zero without causing them to be exactly zero.

These regularization techniques are essential in machine learning, especially when dealing with datasets with a large number of features or when multicollinearity is present. Choosing between Lasso and Ridge depends on the specific characteristics of the data and the desired model behavior.

````python
from sklearn.linear_model import Lasso

X = np.array([[1, 2], [2, 4], [3, 6], [4, 8]])
y = np.array([3, 7, 9, 11])
````

````python
# Fit Lasso Regression model
lasso_model = Lasso(alpha=0.1)
lasso_model.fit(X, y)

# Coefficients and Intercept
coefficients = lasso_model.coef_
intercept = lasso_model.intercept_

print(f"Coefficients: {coefficients}")
print(f"Intercept: {intercept}")
````
